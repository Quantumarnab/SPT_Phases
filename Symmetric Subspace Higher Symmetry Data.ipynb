#  Rounding off to nearest even and odd integers

```julia
function round_odd(N)
   val=Int64(round(N))
    if (val%2)==0     
     if (N-val)>=0
        val+=1
     else
        val-=1
    end
    end
   return val     
end

function round_even(N)
   val=Int64(round(N))
    if (val%2)==1    
     if (N-val)>0
        val+=1
     else
        val-=1 
     end
    end
   return val     
end
```

```julia
using ITensors



function U1(T::MPS,N) 
   
   start=2
   finish=N
   mps=copy(T);
   val=copy(T);

   s = siteind(mps,start)
   newpsi= 2*op(s,"Sz")*mps[start]
   noprime!(newpsi)
   mps[start]= newpsi
    
    
    
    for o = (start+1):6:(finish-2)
           s = siteind(mps,o)
           newpsi= 2*op(s,"Sx")*mps[o]
           noprime!(newpsi)
           mps[o]= newpsi
    end
    
    for m = (start+5):6:(finish-2)
           s = siteind(mps,m)
           newpsi= 2*op(s,"Sx")*mps[m]
           noprime!(newpsi)
           mps[m]= newpsi
    end

   s = siteind(mps,finish-1)
   newpsi= 2*op(s,"Sx")*mps[finish-1]
   noprime!(newpsi)
   mps[finish-1]= newpsi
    
    
    
return mps
end


function U2(T::MPS,N) 
   
   start=1;
   finish=N;
   mps=copy(T);
   val=copy(T);
    
   s = siteind(mps,start)
   newpsi= 2*op(s,"Sz")*mps[start]
   noprime!(newpsi)
   mps[start]= newpsi
    
   t = siteind(mps,start+1)
   newpsi= 2*op(t,"Sx")*mps[start+1]
   noprime!(newpsi)
   mps[start+1]= newpsi
    
    for o = (start+5):6:(finish-2)
           s = siteind(mps,o)
           newpsi= 2*op(s,"Sx")*mps[o]
           noprime!(newpsi)
           mps[o]= newpsi
    end
    
    for m = (start+7):6:(finish-2)
           s = siteind(mps,m)
           newpsi= 2*op(s,"Sx")*mps[m]
           noprime!(newpsi)
           mps[m]= newpsi
    end
    
   u = siteind(mps,finish-1)
   newpsi= 2*op(u,"Sz")*mps[finish-1]
   noprime!(newpsi)
   mps[finish-1]= newpsi
    
   s = siteind(mps,finish)
   newpsi= 2*op(s,"Sx")*mps[finish]
   noprime!(newpsi)
   mps[finish]= newpsi
    
    
    
return mps
end


  function U3(T::MPS,N) 
   
   start=1;
   finish=N;
   mps=copy(T);
   val=copy(T);
    
   s = siteind(mps,start)
   newpsi= 2*op(s,"Sx")*mps[start]
   noprime!(newpsi)
   mps[start]= newpsi
    
    
    
    for o = (start+4):6:(finish-2)
           s = siteind(mps,o)
           newpsi= 2*op(s,"Sx")*mps[o]
           noprime!(newpsi)
           mps[o]= newpsi
    end
    
    
    for m = (start+6):6:(finish-2)
           s = siteind(mps,m)
           newpsi= 2*op(s,"Sx")*mps[m]
           noprime!(newpsi)
           mps[m]= newpsi
    end
    

   s = siteind(mps,finish)
   newpsi= 2*op(s,"Sz")*mps[finish]
   noprime!(newpsi)
   mps[finish]= newpsi

    
return mps
end


function U4(T::MPS,start,finish) 
   
   start=1;
   finish=N;
   mps=copy(T);
   val=copy(T);
    
   s = siteind(mps,start)
   newpsi= 2*op(s,"Sz")*mps[start]
   noprime!(newpsi)
   mps[start]= newpsi
    
    
    for o = (start+3):6:(finish-2)
           s = siteind(mps,o)
           newpsi= 2*op(s,"Sx")*mps[o]
           noprime!(newpsi)
           mps[o]= newpsi
    end
    
    for o1 = (start+5):6:(finish-2)
           s = siteind(mps,o1)
           newpsi= 2*op(s,"Sx")*mps[o1]
           noprime!(newpsi)
           mps[o1]= newpsi
    end
    
    
   v = siteind(mps,finish)
   newpsi= 2*op(v,"Sx")*mps[finish]
   noprime!(newpsi)
   mps[finish]= newpsi
    
    
    
return mps
end
```
